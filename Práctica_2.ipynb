```python
# Este codigo es el correspondiente para contestar las preguntas uno y dos de
# la practica, para modificar el tiempo de atencion, se escribio la opcion de
# tomar el tiempo de la consulta en 20 minutos como un comentario en este mismo
# codigo, por lo que solo habria que cambiar el "gato" para que funcionara.

import random as rd
from collections import deque

class Paciente():
  '''Representa una persona que asiste al hospital.'''
  def __init__(self, numeroP):
    '''Inicializa a un nuevo paciente con distintos atributos por definir.'''
    self.numeroP = numeroP # Numero unico de cada paciente.
    self.Tipo = self.Tipo() # Tipo de paciente, Urgente o Rutinario.
    self.horaEntrada = None # Hora de entrada del paciente a la clinica.
    self.horaEntradaConsultorio = None
    # Hora de entrada del paciente entrada del paciente al consultorio.

    self.horaSalida = None
    # Hora de salida del paciente del consultorio y de la clinica.

    self.tiempoEspera = None
    # Tiempo de espera del paciente para entrar al consultorio.

    self.tiempoSistema = None
    # Tiempo que el paciente estuvo en el sistema de la clinica.

  def Tipo(self):
    '''Determina si se trata de un paciente de Urgencia o Rutinario.'''
    prob = rd.uniform(0, 100)
    if prob <= 20:
        return 'Urgente'
    else:
        return 'Rutinario'

class Cola():
  '''Representa una cola.'''
  def __init__(self):
    '''Inicializa una cola vacia.'''
    self.lista = deque() # Cola de la llegada de los pacientes a la clinica.
    self.espera = deque() # Cola de pacientes en espera.
    self.consultorio = deque() # Cola de pacientes en el consultorio.

  def Ingreso(self, paciente):
    '''Agrega a un paciente a la lista previa antes de pasar al consultorio.
    Crea la existencia de los pacientes antes de entrar a la clínica.'''
    self.lista.append(paciente)
    return self.lista

  def Espera(self):
    '''Agrega a un paciente de la lista anterior a la cola de espera.
    Simula la sala de espera de la clínica.'''
    paciente = self.lista.popleft()
    self.espera.append(paciente)
    return self.lista, paciente

  def Consultorio(self):
    '''Agrega a un paciente de la cola de espera a la cola de consultorio.
    Simula el consultorio de la clinica.'''
    paciente = self.espera.popleft()
    self.consultorio.append(paciente)
    return self.espera, paciente

  def Egreso(self):
    '''Elimina a un paciente de la cola de consultorio.
    Simula la partida del paciente despues de terminar su consulta.'''
    paciente = self.consultorio.popleft()
    return self.consultorio, paciente

class Simulacion():
  '''Representa la simulación del sistema de atención médica de la clínica. '''
  def __init__(self):
    '''Inicializa la simulación, estableciendo el tiempo en cero.'''
    self.Minutos = 0

  def Reloj(self, minutos):
    '''Representa/simula el funcionamiento de un reloj.'''
    self.Minutos += round(minutos)

Cola = Cola()
Simulacion = Simulacion()
paciente_N = None
# Esto para la parte del Consultorio y no perder a los pacientes.

paciente_C = None # Variable para paciente en consulta.
paciente_O = None # Variable para paciente que termino su consulta.
numeroP = 1 # Paciente 1.

# Simucion de la llegada de todos los pacientes.
while numeroP <= 50000:
  interarribo = rd.uniform(0.5, 50)
  # Genera tiempos aleatorios entre las llegadas de los pacientes a la clinica.

  Simulacion.Reloj(interarribo) # Avanza el tiempo de la simulacion.
  paciente = Paciente(numeroP) # Se crea a un nuevo paciente.
  Cola.Ingreso(paciente) # Agrega al paciente a la cola de ingreso a la clinica.
  paciente.horaEntrada = Simulacion.Minutos
  # Registra las horas de entrada de los pacientes.

  numeroP += 1 #Incrementa el numero de los pacientes.

TiemposEspera = []
# Lista para almacenar los tiempos de espera de los pacientes.

TiemposSistema = []
# Lista donde se almacenan los tiempos en el sistema de los pacientes.

Simulacion.Minutos = 0 # Reinicia el tiempo de la simluacion.

# Simulacion del funcionamiento de la clinica.
while True:
  # Si hay pacientes en la cola de llegada y si es su momento de llegada, se
  # mueven a la cola de espera.
  if Cola.lista and Cola.lista[0].horaEntrada == Simulacion.Minutos:
      print(f"Ingresa el Paciente {Cola.lista[0].numeroP} de tipo {Cola.lista[0].Tipo} a Sala de Espera a los {Cola.lista[0].horaEntrada} minutos.")
      paciente_N = Cola.Espera()[1] # Mueve al paciente a la cola de espera.
      print("")

# Si hay pacientes en la cola de espera y es su turno de ser atendidos, se
# mueven a la cola consultorio.
  if Cola.espera and Cola.espera[0].horaEntrada <= Simulacion.Minutos:
    if not Cola.consultorio:
      # Verifica si no hay pacientes en la cola consultorio.

      paciente_C = Cola.Consultorio()[1]
      # Mueve al paciente a la cola consultorio.

# Si el paciente esta en la cola consultorio, se registra su tiempo de espera y
# se calcula su salida de esta cola (su salida de la sala de espera).
  if paciente_C is not None:
    paciente_C.tiempoEspera = Simulacion.Minutos - paciente_C.horaEntrada
    paciente_C.horaEntradaConsultorio = Simulacion.Minutos
    #paciente_C.horaSalida = paciente_C.horaEntradaConsultorio + 30
    # Duracion de la consulta.

    paciente_C.horaSalida = paciente_C.horaEntradaConsultorio + 20
    # Duracion de la consulta.

    print(f"El Paciente {paciente_C.numeroP} de tipo {paciente_C.Tipo} entra al consultorio a los {paciente_C.horaEntradaConsultorio} minutos después de esperar {paciente_C.tiempoEspera} minutos.")
    paciente_C = None

# Si es hora de salida de un paciente de la cola consultorio,
# se saca de esta cola.
  if Cola.consultorio and Cola.consultorio[0].horaSalida == Simulacion.Minutos:
    paciente_O = Cola.Egreso()[1] # Elimina al paciente de la cola consultorio.

# Si el paciente termino su consulta, se registra su tiempo en el sistema y
# se imprime su salida.
  if paciente_O is not None:
    paciente_O.tiempoSistema = paciente_O.horaSalida - paciente_O.horaEntrada
    print(f"El Paciente {paciente_O.numeroP} de tipo {paciente_O.Tipo} sale del consultorio a los {paciente_O.horaSalida} minutos.")
    print(f"El Paciente {paciente_O.numeroP} ha estado {paciente_O.tiempoSistema} minutos en la clínica")
    TiemposEspera.append(paciente_O.tiempoEspera)
    TiemposSistema.append(paciente_O.tiempoSistema)
    paciente_O = None
    print("")

  Simulacion.Minutos += 1 #Avance del tiempo en la simulacion.

  #Si no hay pacientes en ninguna de las tres colas, finaliza la simulacion.
  if not Cola.lista and not Cola.espera and not Cola.consultorio:
    break

# Calcula el tiempo promedio de espera y en el sistema de los pacientes.
Suma_S = 0
Suma_E = 0
for i in TiemposSistema:
  Suma_S += i

for i in TiemposEspera:
  Suma_E += i

Promedio_S = round(Suma_S / len(TiemposSistema))
Promedio_E = round(Suma_E / len(TiemposEspera))
print(f"El promedio de los tiempos de espera es de {Promedio_E} minutos.")
print(f"El promedio de los tiempos en el sistema es de {Promedio_S} minutos.")
```
